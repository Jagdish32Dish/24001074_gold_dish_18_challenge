```python
import pandas as pd

file_path = 'data.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')


df_filtered = df[(df['HS_Weak'] == 1) | (df['HS_Moderate'] == 1) | (df['HS_Strong'] == 1)]
df_filtered = df_filtered[['Tweet', 'HS_Weak', 'HS_Moderate', 'HS_Strong']]

print(df_filtered)
```

```python
import pandas as pd
import matplotlib.pyplot as plt



df_filtered = df_filtered[['Tweet', 'HS_Weak', 'HS_Moderate', 'HS_Strong']]


total_weak = df_filtered['HS_Weak'].sum()
total_moderate = df_filtered['HS_Moderate'].sum()
total_strong = df_filtered['HS_Strong'].sum()


data = {'HS_Weak': total_weak, 'HS_Moderate': total_moderate, 'HS_Strong': total_strong}


series = pd.Series(data)


plt.figure(figsize=(8, 8))
plt.pie(series, labels=series.index, autopct='%1.1f%%', startangle=140)
plt.title('Percentage Distribution of HS_Weak, HS_Moderate, HS_Strong')
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

file_path = 'data.csv'
df = pd.read_csv(file_path, encoding='ISO-8859-1')


df_filtered = df[(df['HS_Weak'] == 1) | (df['HS_Moderate'] == 1) | (df['HS_Strong'] == 1)]


df_filtered = df_filtered[['Tweet', 'HS_Individual', 'HS_Group']]


counts = df_filtered[['HS_Individual', 'HS_Group']].apply(pd.Series.value_counts).fillna(0).astype(int).T


sns.set(style="whitegrid")
plt.figure(figsize=(12, 6))
counts.plot(kind='bar', stacked=True, colormap='viridis', ax=plt.gca())
plt.title('Counts of "1" in HS_Individual and HS_Group')
plt.xlabel('HS Type')
plt.ylabel('Count')
plt.show()
```
